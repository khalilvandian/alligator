# Filter Combined Datasets For organisations

```python
org_subclasses_set = set(org_subclasses)

def check_if_is_org(x):
    classes_set = set(entity_class[0] for entity_class in x if len(entity_class) > 0)
    org_classes = classes_set.intersection(org_subclasses_set)
    percentage = len(org_classes)/len(x) if len(x) != 0 else 0

    if percentage >= 0.5:
        return True
    else:
        return False

filtered_datasets = combined_classified_datasets[combined_classified_datasets[4].progress_apply(check_if_is_org)]
# WikidataTables2023R1["classes"].progress_apply(check_if_is_org)
```

## Manual Filtering using a set, to exclude cities, districts, villages, ...

```python
import random

def filter_classes(x):

    # The filter for unwanted entity classes
    filter = {"village", "district", "city", "municipality", "province", "U.S. state", "state of Venezuela", "commune of",
              "rural settlement", "urban settlement", "town", "township", "municipality of the Philippines", "municipality of Colombia",
              "legislative term", "prefecture of Japan", "country", "region of", "state of", "human settlement", "area", "reserve in"}

    if len(x) == 0:
        return True
    else:
        for entity_class in x:
            for filt in filter:
                # if random.random() < 0.02:
                #     print(entity_class[1], filt)

                if filt in entity_class[1]:
                    # if random.random() < 0.05:
                    #     print("Class: ", entity_class[1], " - Filter: ", filt)
                    return False
                
    return True

mask = filtered_datasets[4].progress_apply(filter_classes)
filtered_datasets_2 = filtered_datasets[mask]
filtered_rows = filtered_datasets[~mask]

# sample the data with seed = 100
# random.seed(100)
# print(filtered_datasets_2.sample(50).iloc[:, [3, 4]].to_string())

# print("length is: ", len(filtered_datasets[filtered_datasets[4].apply(filter_classes)]))
```
