# Filter Combined Datasets For organisations

```python
org_subclasses_set = set(org_subclasses)

def check_if_is_org(x):
    classes_set = set(entity_class[0] for entity_class in x if len(entity_class) > 0)
    org_classes = classes_set.intersection(org_subclasses_set)
    percentage = len(org_classes)/len(x) if len(x) != 0 else 0

    if percentage >= 0.5:
        return True
    else:
        return False

filtered_datasets = combined_classified_datasets[combined_classified_datasets[4].progress_apply(check_if_is_org)]
# WikidataTables2023R1["classes"].progress_apply(check_if_is_org)
```
